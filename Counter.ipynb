```python
import pandas as pd
```

```python
begin = pd.read_csv("data/output.txt", header = None)
```

```python
grouped = begin.groupby([0]).size().sort_values(ascending=False)
grouped = grouped.reset_index(name="counts")
grouped
```

```python
top100k = grouped.iloc[range(100000),:]
top100k['words'] = top100k[0].str.split().str.len()
```

```python
top100k[top100k['words']==2].nlargest(10, 'counts')
```

```python
top100k[top100k['words']==3].nlargest(10, 'counts')
```

```python
top100k[top100k['words'] == top100k['words'].max()]
```

```python
top100k[top100k[0]==""]
```
